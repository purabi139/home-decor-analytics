```python
import pandas as pd

# Generate a dataset for current market trends and top competitors in the home decor and lifestyle sector

# Define competitor names and their strategies
competitors = [
    {"Name": "Wayfair", "Strategy": "Discount Pricing, Free Shipping, Online Exclusives"},
    {"Name": "IKEA", "Strategy": "Flat-Pack Furniture, Affordable Prices, In-Store Experience"},
    {"Name": "West Elm", "Strategy": "Modern Designs, Sustainable Sourcing, Premium Pricing"},
    {"Name": "Pottery Barn", "Strategy": "Classic Designs, Seasonal Collections, Loyalty Programs"},
    {"Name": "Amazon", "Strategy": "Wide Range, Quick Delivery, Competitive Pricing"},
    {"Name": "Etsy", "Strategy": "Handmade Items, Unique Designs, Artisanal Appeal"}
]

time_periods = pd.date_range(start="2023-01-01", end="2023-12-31", freq="M")



```

```python
import random

categories = [
    "Minimalist Furniture", "Eco-Friendly Decor", "Boho Style", 
    "Smart Home Devices", "Wall Art", "Sustainable Lighting", 
    "Vintage Decor", "Scandinavian Design", "Custom Furniture"
]
market_trends_data = []
for competitor in competitors:
    for date in time_periods:
        category = random.choice(categories)
        market_share = round(random.uniform(5, 40), 2)  # Random market share percentage
        customer_rating = round(random.uniform(3.5, 5.0), 2)  # Random customer rating
        advertising_spend = round(random.uniform(5000, 100000), 2)  # Advertising spend in USD
        trend_growth = round(random.uniform(0.05, 0.2), 2)  # Growth in trend popularity

        market_trends_data.append({
            "Competitor": competitor["Name"],
            "Strategy": competitor["Strategy"],
            "Category": category,
            "Date": date,
            "MarketShare": market_share,
            "CustomerRating": customer_rating,
            "AdvertisingSpend": advertising_spend,
            "TrendGrowth": trend_growth
        })
```

```python
# Convert to a DataFrame
competitor_trends_df = pd.DataFrame(market_trends_data)

# Save the dataset to a CSV file
competitor_file_path = "/Users/home/Downloads/Home_Decor/Competitor_Market_Trends.csv"
competitor_trends_df.to_csv(competitor_file_path, index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
competitor_file_path = "/Users/home/Downloads/Home_Decor/Competitor_Market_Trends.csv"
data = pd.read_csv(competitor_file_path)

# Top competitors by average market share
top_competitors = data.groupby("Competitor")["MarketShare"].mean().sort_values(ascending=False)
```

```python
# Competitor strategies and market share visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=top_competitors.index, y=top_competitors.values, palette="viridis")
plt.title("Top Competitors by Average Market Share ", fontsize=16)
plt.xlabel("Competitor", fontsize=12)
plt.ylabel("Average Market Share", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python



# Advertising budget trends over time for top competitors
top_competitors_list = top_competitors.index[:5]  # Select top 5 competitors
filtered_data = data[data["Competitor"].isin(top_competitors_list)]

plt.figure(figsize=(14, 7))
sns.barplot(data=filtered_data, x="AdvertisingSpend", y="Date",hue="Competitor")
#sns.lineplot(data=filtered_data, x="Date", y="AdvertisingSpend", hue="Competitor", palette="tab10")
plt.title("Advertising Budget Trends Over Time (Top Competitors)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("AdvertisingSpend", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Competitor")
plt.show()
```

```python
# Market trends by growth rate across categories
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x="Category", y="TrendGrowth", palette="coolwarm")
plt.title("Market Trends by Growth Rate Across Categories", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("TrendGrowth", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python

# Customer rating distribution by competitor
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x="Competitor", y="CustomerRating", palette="pastel")
plt.title("Customer Ratings Distribution by Competitor", fontsize=16)
plt.xlabel("Competitor", fontsize=12)
plt.ylabel("Customer Rating", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Advertising budget trends over time for top competitors
top_competitors_list = top_competitors.index[:3]  # Select top 5 competitors
filtered_data = data[data["Competitor"].isin(top_competitors_list)]

plt.figure(figsize=(14, 7))
sns.barplot(data=filtered_data, x="Date", y="AdvertisingSpend", hue="Competitor", palette="tab10")
plt.title("Advertising Budget Trends Over Time (Top Competitors)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Advertising Spend", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Competitor")
plt.show()
```
